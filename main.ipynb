```python
from diagrams import Cluster, Diagram
from diagrams.custom import Custom
from diagrams.onprem.client import Client, User, Users
from diagrams.onprem.compute import Server
from diagrams.generic.device import Mobile
from diagrams.onprem.database import Postgresql
from diagrams.onprem.inmemory import Redis
from diagrams.saas.chat import Discord, Slack

def Phoenix(name):
    return Custom(name, "./logo/Phoenix.png")
def Shopify(name):
    return Custom(name, "./logo/Shopify.png")
def Magento(name):
    return Custom(name, "./logo/Magento.png")
def Woo(name):
    return Custom(name, "./logo/Woo.png")

with Diagram("Phi Suite", show=False, direction="LR") as diag:
    with Cluster("External"):
        with Cluster("Services"):
            shopify = Shopify("Shopify")
            magento = Magento("Magento")
            woo = Woo("WooCommerce")
            discord = Discord("Discord")
            slack = Slack("Slack")
        with Cluster("IoT Devices"):
            mobile = Mobile("Mobile")
            client = Client("Computer")
            server = Server("Server")   
    with Cluster("Event-Driven Data Engineering Platform"):
        with Cluster("Super Users"):
            admin = User("Administrator")
        with Cluster("Data Team X"):
            customers = Users("Data Scientists")
        with Cluster("Data Team Y"):
            operators = Users("Data Scientists")
        with Cluster("Monitoring Team"):
            qa = Users("Quality Assurance")
        
        with Cluster("Schema App"):
            with Cluster("Phoenix Server"):
                a_ui = Phoenix("Admin LiveView")
                a_api = Phoenix("Admin API")
            with Cluster("Schema Store"):
                a_db = Postgresql("Prostgres")
        with Cluster("Event Schema MQ"):
            ac_mq = Redis("Redis PubSub")
        with Cluster("Entity Schema MQ"):
            aq_mq = Redis("Redis PubSub")
        with Cluster("Process Schema MQ"):
            ap_mq = Redis("Redis PubSub")
                
        with Cluster("Event App"):
            with Cluster("Phoenix Server"):
                c_ui = Phoenix("Event LiveView")
                c_api = Phoenix("Event API")                
            with Cluster("Event Store"):
                c_db = Postgresql("Prostgres")
        with Cluster("Event Data MQ"):
            c_mq = Redis("Redis PubSub")
                
        with Cluster("Entity App"):
            with Cluster("Phoenix Server"):
                q_ui = Phoenix("Enitity  LiveView")
                q_api = Phoenix("Entity API")             
            with Cluster("Entity Store"):
                q_db = Postgresql("Postgres")                
                
        with Cluster("Entity Data MQ"):
            q_mq = Redis("Redis PubSub")
                
        with Cluster("Process App"):
            with Cluster("Phoenix Server"):
                p_ui = Phoenix("Process LiveView")
                p_api = Phoenix("Process API")             
            with Cluster("Process Store"):
                p_db = Postgresql("Postgres")                
    
    a_ui >> a_api >> a_db >> a_api >> [a_ui, ac_mq, aq_mq, ap_mq]
    ac_mq >> c_api
    aq_mq >> q_api
    ap_mq >> p_api
    c_api >> c_db >> c_api >> [c_ui, c_mq]
    c_mq >> p_api
    q_ui >> q_api >> q_db >> q_api >> [q_ui, q_mq]
    q_mq >> c_api
    p_api >> p_db >> p_api >> p_ui
    server - client - mobile - magento - woo - shopify - discord - slack
    admin >> a_ui >> admin
    customers >> q_ui >> customers
    # client >> c_api
    shopify >> c_api
    p_api >> slack
    discord >> p_api

    qa << p_ui
    operators << c_ui
    
diag
```

```python
from diagrams import Cluster, Diagram
from diagrams.custom import Custom
from diagrams.onprem.client import User, Users
from diagrams.programming.language import Typescript

def Svelte(name):
    return Custom(name, "./logo/Svelte.png")
def Ethers(name):
    return Custom(name, "./logo/Ethers.png")
def Hardhat(name):
    return Custom(name, "./logo/Hardhat.png")
def Vyper(name):
    return Custom(name, "./logo/Vyper.png")
def OpenZeppelin(name):
    return Custom(name, "./logo/OpenZeppelin.png")

with Diagram("FairChest", show=False, direction="LR") as diag:
    
    with Cluster("Web3 Application"):
        with Cluster("Users"):
            admin = User("Administrator")
            client = Users("Clients")
            contractor = Users("Contractors")
        with Cluster("Web Interface"):
            svelte = Svelte("SvelteKit")
            ethers = Ethers("Ethers")
        with Cluster("Development Environment"):
            hardhat = Hardhat("Hardhat")
            ts = Typescript("TypeScript")
        with Cluster("Ethereum Blockchain"):
            with Cluster("Smart Contracts"):
                chest = Vyper("FairChest")
                with Cluster("ERC721"):
                    doc = OpenZeppelin("Documents")
                    req = OpenZeppelin("Requests")
                    pro = OpenZeppelin("Proposals")
                    rep = OpenZeppelin("Reports")
    
    ts - hardhat
    svelte >> ethers >> chest >> [doc, req, pro, rep]
    svelte << ethers << chest
    admin - client - contractor
    svelte >> client >> svelte
    hardhat >> [ethers, chest]
    
diag
```
